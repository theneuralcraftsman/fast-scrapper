```python
import re
```

```python
def extract_hrefs(html_content):
    # Define the regular expression pattern to match the href attribute
    pattern = r'href="([^"]*/case/[^"]*)"'

    # Use re.findall to find all matches of the pattern in the HTML content
    hrefs = re.findall(pattern, html_content)

    # Join the list of href values using commas
    hrefs_string = '\", \"https://casetext.com'.join(hrefs)

    return hrefs_string
```

```python
def extract_hrefs_as_list(html_content):
    # Define the regular expression pattern to match the href attribute
    pattern = r'href="([^"]*/case/[^"]*)"'

    # Use re.findall to find all matches of the pattern in the HTML content
    hrefs = re.findall(pattern, html_content)

    # Prepend "https://casetext.com" to each href
    full_hrefs = ['https://casetext.com' + href for href in hrefs]

    return full_hrefs
```

```python
from bs4 import BeautifulSoup as Soup

n = 1
new_href_values=''


while n<=12:
  n = n+1
  url = f"https://casetext.com/cases/us/2023/{n}"
  loader = RecursiveUrlLoader(
      url=url, max_depth=3
  )
  new_doc_temp = loader.load()
  # Extract href values
  new_href_values += extract_hrefs(new_doc_temp[0].page_content)

```

```python
def get_page_data(url):
  from bs4 import BeautifulSoup as Soup

  url = url
  loader = RecursiveUrlLoader(
      url=url, max_depth=2, extractor=lambda x: Soup(x, "html.parser").text
  )
  page_content_required = loader.load()
  return page_content_required
```

```python
from bs4 import BeautifulSoup as Soup

n = 1
new_href_values=[]

year = 2013

while year <= 2023:
    n = 1  # Reset month counter for each year
    while n <= 12:
        url = f"https://casetext.com/cases/us/{year}/{n}"
        loader = RecursiveUrlLoader(
            url=url, max_depth=3
        )
        new_doc_temp = loader.load()
        # Extract href values
        # Extract href values as a list and extend new_href_values list
        new_href_values.extend(extract_hrefs_as_list(new_doc_temp[0].page_content))

        # Print dynamic progress indicator
        print(f"\rProcessing year: {year}, month: {n}", end='', flush=True)

        n += 1  # Move to the next month

    year += 1  # Move to the next year

print("\nCompleted!")
# print(new_href_values)
```
