```python
import re
```

```python
def extract_hrefs(html_content):
    # Define the regular expression pattern to match the href attribute
    pattern = r'href="([^"]*/case/[^"]*)"'

    # Use re.findall to find all matches of the pattern in the HTML content
    hrefs = re.findall(pattern, html_content)

    # Join the list of href values using commas
    hrefs_string = '\", \"https://casetext.com'.join(hrefs)

    return hrefs_string
```

```python
def extract_hrefs_as_list(html_content):
    # Define the regular expression pattern to match the href attribute
    pattern = r'href="([^"]*/case/[^"]*)"'

    # Use re.findall to find all matches of the pattern in the HTML content
    hrefs = re.findall(pattern, html_content)

    # Prepend "https://casetext.com" to each href
    full_hrefs = ['https://casetext.com' + href for href in hrefs]

    return full_hrefs
```

```python
def extract_hrefs_for_mid_links(html_content, cases):
    # Define the regular expression pattern to match the href attribute
    pattern = rf'href="([^"]*/{cases}/[^"]*)"'

    # Use re.findall to find all matches of the pattern in the HTML content
    hrefs = re.findall(pattern, html_content)

    # Prepend "https://casetext.com" to each href
    full_hrefs = ['https://casetext.com' + href for href in hrefs]

    return full_hrefs
```

```python
from bs4 import BeautifulSoup as Soup

n = 1
mid_links_list=[]

cases = "wi"

while n<=12:
  n = n+1
  url = f"https://casetext.com/cases/{cases}/"
  loader = RecursiveUrlLoader(
      url=url, max_depth=1
  )
  new_doc_temp = loader.load()
```

```python
# Extracting keywords

mid_link_key = []

for url in mid_links_list:
    # Split the URL by "/"
    parts = url.split("/")
    # Extract the part containing the keyword
    keyword = parts[-1]
    # Append the keyword to the list
    mid_link_key.append(keyword)
```

```python
## If there are mid links
import requests
from requests.exceptions import RequestException
from bs4 import BeautifulSoup as Soup
import time

n = 1
new_href_values=[]

year = 2018
delay_between_retry_requests = 5

for mid_element in mid_link_key:

    year = 2018  # Reset year for each mid_element
    while year <= 2023:
        n = 1  # Reset month counter for each year
        while n <= 12:

            url = f"https://casetext.com/cases/{mid_element}/{year}/{n}"
            max_retries = 3  # Set the number of retries
            retry_count = 0

            delay_between_retry_requests=5  # Resetting retry timer

            for retry_count in range(max_retries):
                  loader = RecursiveUrlLoader(
                      url=url, max_depth=1
                  )
                  new_doc_temp = loader.load()
                  if len(new_doc_temp) == 0:
                    print(f"Waiting for {delay_between_retry_requests} secs and retrying...")
                    time.sleep(delay_between_retry_requests)
                    delay_between_retry_requests+=5
                    retry_count+=1
                  else:
                    break




            # Extract href values
            # Extract href values as a list and extend new_href_values list
            new_href_values.extend(extract_hrefs_as_list(new_doc_temp[0].page_content))

            # Print dynamic progress indicator
            print(f"\rProcessing Mid Link: {mid_element} year: {year}, month: {n}", end='', flush=True)
            #print(f"Year: {year} Month: {n}")
            n += 1  # Move to the next month
            time.sleep(2)   # Wait for 2 secs before going to next year

        year += 1  # Move to the next year
        time.sleep(5) # Wait for 5 secs before going to next year

    time.sleep(10) # Wait for 10 seconds before trying next mid link


print("\nCompleted!")
```
